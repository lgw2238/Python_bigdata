```python
'''
# barchar는 bar(세로 방향), barh(가로방향)
# bar(x,y) : x축, y축

'''
```

```python
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
```

```python
y = [2,3,1]
x = np.arange(3)


plt.bar(x,y)
plt.grid(True)
plt.show()
```

```python
y = [2,3,1]
x = np.arange(3)


plt.barh(x,y)
plt.grid(True)
plt.show()
```

### histogram

```python
'''
# hist(bins=)  #bins: 데이터를 집계할 구간 정보


'''
```

```python
np.random.seed(0)
x = np.random.randn(1000)
plt.hist(x, bins=10)
plt.show()
```

## pie

```python
'''
# pie(ratio[비율], explode[간격], labels, colors, autopot[소수점 자리수], shadow, startangle[시작각])
)
원의 형태 유지 : plt.axis('equal')

'''
```

```python
labels = 'A', 'B', 'C', 'D'

ratio = [10, 30, 40 , 20]
colors = ['red', 'skyblue', 'yellowgreen', 'pink']
explode = (0, 0.1, 0, 0)

plt.pie(ratio, explode=explode, labels=labels, colors=colors, shadow=True, startangle=90,
        autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

## scatter plot





```python
'''
# 스캐터 플롯(산점도 함수) : 두 개의 실수 데이터 집합의 상관관계를 살펴 볼 때 많이 사용하는 차트
# scatter(x,y)

'''
```

```python
# 정규 분포
m, sigma = 0, 0.1
s = np.random.normal(m, sigma, 1000)


plt.hist(s, 30)
plt.grid(True)
plt.show()
```

```python
# scatter
data1=[5, 9, 11, -3, 7, -3, 13, 2,  8]
data2=[9, 11, 3, 6, -3, 13, 5,  8, 9]
plt.figure()

plt.scatter(data1, data2, s=53, c="red", alpha=0.5,  label='Inline label' )
plt.scatter([1,2,3], [3,5,9], s=53, c="blue", alpha=0.5,  label='Inline label' )
plt.xlim([-10,30])
plt.ylim([-10,30])
#plt.autoscale(enable=True, axis='y')

plt.legend()
plt.show()

# np.random.normal(평균, 표준편차, 샘플 수)
```

## bubble chart

```python
'''
# 데이터가 2차원이 아닌 3차원 이상의 경우 점 하나의 크기 또는 컬러를 이용하여 표현
# 크기 : s, 컬러 : c 

'''
```

```python
n = 30 
np.random.seed(0)
x= np. random.rand(n)
y1 = np.random.rand(n)
y2 = np.random.rand(n)
y3 = np.pi * (15 * np.random.rand(n))**2


plt.scatter(x, y1, c=y2, s=y3)
plt.grid()
plt.show()
```
