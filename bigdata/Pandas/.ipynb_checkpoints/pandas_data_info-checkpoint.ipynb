### 데이터 개수 확인 

```python
# 데이터 프레임 각 열에 대해서 원소 개수 확인
df.count()
```

```python
# 특정 열이 가지고 있는 고유값 확인
# origin : 제조국(ex: USA - 1, UK - 2..)
df['origin'].value_counts()
# 고유값이 행 인덱스, 고유값의 갯수가 데이터 값
```
